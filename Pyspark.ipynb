สร้าง Spark Session

```python
from pyspark.sql import SparkSession

# สร้าง SparkSession
spark = SparkSession.builder \
    .appName("MyFirstPySparkApp") \
    .getOrCreate()

print(spark)
```

```python
df = spark.read.csv("preprocessed_data.csv", header=True, inferSchema=True)
df.show(5) 
```

```python
df.printSchema()
```

```python
df.count()
```

```python
df.select(['Biomass_Boiler_Actual_Oxygen',"Biomass_Boiler_Pressure_Boiler","Biomass_Boiler_Steam_Flow_Rate"]).show()
```

```python
df[['Biomass_Boiler_Temperature_Furnace']].filter(df["Biomass_Boiler_Actual_Oxygen"] > 10).show()
```

ทำงานกับ DataFrame แบบ SQL

```python
df.createOrReplaceTempView("boiler")
```

```python
spark.sql("""SELECT Biomass_Boiler_Temperature_Furnace 
          FROM boiler 
          WHERE Biomass_Boiler_Actual_Oxygen > 10 
          ORDER BY Biomass_Boiler_Temperature_Furnace DESC""").show()
```

```python
# สร้าง DataFrame จาก Pandas DataFrame
pandas_df = df.toPandas()
```
